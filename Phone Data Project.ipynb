## Project Objectives
- To find out which brand of phones that customers prefer
- To find out the regions had the highest phone sales

## Example of the Dataset being worked on


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Data=pd.read_csv("SarahFinal.csv")
Data.head(10)
```

```python
#Data.drop_duplicates()
Data.isnull().any() # columns with null values
#Data.head(10)
```

```python
#Data.info()
#Data.info()
#Data_dropped=pd.DataFrame(Data.drop(Data.columns[[1,4,10,11]], axis=1, inplace=True))
#Data_dropped.head()
Data_drop=pd.DataFrame(Data.iloc[:, [0,2,3,5,6]])
print(Data_drop.head())
```

# To find out which teams sold the highest phone sales

```python
#Answering Objective three-To Find Out Which Teams Sold The Highest Phone Sales
team_numbers=Data_drop['Team'].value_counts()
print(team_numbers)
```

```python
team_numbers=(Data_drop['Team'].value_counts()).to_numpy()
print(np.sum(team_numbers))
print(team_numbers)
plt.pie(team_numbers,labels=["Black","Purple","Orange","White","Brown","Maroon"],autopct="%1.2f%%",
       explode=[0.2,0,0,0,0.2,0.2])
plt.show()
# Black contributed the highest phone sales compared to Maroon and Brown which contributed the lowest phone sales
```

# Finding out which brand of phones customers prefer

```python
#Answering Objective one-To Find Out Which Brand Of Phones Customers Prefer
phonemodel_numbers=Data_drop['Plan Sold'].value_counts()
print(phonemodel_numbers)
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.gca()
Data_drop['Plan Sold'].hist(grid=False, edgecolor='black', ax=ax,bins=8)
plt.xticks(rotation = 45)
plt.show()
```

# Regions with the highest number of phone sales

```python
#Answering Objective Two-To Find Out Regions That Had The Highest Number Of Phone Sales
Region=Data_drop['Town'].value_counts()
print(Region)
```

```python
Region=(Data_drop['Town'].value_counts()).to_numpy()
print(np.sum(Region))
```

```python
Data_drop_cut=Data_drop.select_dtypes(["object"]).columns
Data_drop[Data_drop_cut]=Data_drop[Data_drop_cut].apply(lambda x: pd.factorize(x)[0])
Data_dropcorr=Data_drop.corr()
print(Data_dropcorr)
fig=plt.figure(figsize=(8,8))
plot=sns.heatmap(Data_drop.corr().round(2),annot=True,cmap="Purples",mask=np.triu(Data_dropcorr))
plot.set_title("Regions with the most phones sold")
plt.show()
# In summary, sales from agents depends on the region
```

```python
xlist = np.linspace(-10, 10, 100)
ylist = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
Data_drop = ax.contourf(X, Y, Z)
# Add a colorbar to a plot
fig.colorbar(Data_drop) 
ax.set_title('Regions')
plt.show()
```
